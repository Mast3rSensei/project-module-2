# The Unbalancing Dilemma

As we can see from the bar chart, each type of cuisine has quite a different number of restaurants registered. This means that, if we want to calculate an estimator of the "real" population's mean and variance, when taking a random sample of *n* restaurants, we would probably get a biased estimator. What if, out of a random sample of 500 restaurants, we end up with a sample that has 400 Portuguese restaurants and only 100 from different cuisines? This would not be so uncommon considering that portuguese restaurants accounts for almost 23% of the total number of restaurants Zomato has. What we need is a **Stratified Random Sampling**.

Stratified Random Sampling is a technique that will ensure our population's sample will fairly include the right number of variables from each **strata**. These numbers are going to be decided by the following formula, applied to each strata:
<br>
<br>
<div align=center>$n_j = \frac{N_j}{N}\times{n}$</div>

where:

$n_j$ = the size of the sample's strata  
$N_j$ = the size of the population's strata  
$N$ =  the population's size  
$n$ = the sample size

Once obtained the stratified sample, the rest is about applying the right formulas to obtain the estimator for the population's mean and variance, along with the confidence interval.

Suppose that random samples of nj individuals are taken from strata containing $N_j$ individuals $(j = 1, 2, ..., K)$. Let
<br><br>
<div align=center>$\sum_{j=1}^{K} N_j = N$</div> and <div align=center>$\sum_{j=1}^{K} n_j = n$</div>

Denote the sample means and variances in the strata by $x_j$ and $s^2_j (j = 1, 2, ..., K)$ and the overall population mean by $\mu$.

1. An unbiased estimation procedure for the overall population mean $\mu$ yields the following point estimate:
<br><br>
<div align=center>$\bar{x}_{st} = \frac{1}{N}\sum_{j=1}^{K} N_j\bar{x}_j$</div>
<br>
2. An unbiased estimation procedure for the variance of our estimator of the overall population mean yields the point estimate:
<br><br>
<div align=center>$\hat\sigma^2_{\bar{x}_{st}} = \frac{1}{N^2}\sum_{j=1}^{K}N^2_j\hat\sigma^2_{x_j}$</div>
<br>
where
<br><br>
<div align=center>$\hat\sigma^2_{\bar{x_j}} = \frac{s^2_j}{n_j}\times{\frac{N_j - n_j}{N_j - 1}}$</div>
<br>
3. Provided the sample size is large, a 100(1 - $\alpha$)% confidence interval estimation of the population mean using stratified random samples, is obtained from the following:
<br><br>
<div align=center>$\bar{x}_{st}+/-{z_{\alpha/2}\hat\sigma_{\bar{x}_{st}}}$</div>

```python
def stratified_pop_estimator(dataframe, str_variable, num_variable, sample_size):
    strat_count = dataframe.groupby(str_variable)[num_variable].count()
    N = sum(strat_count)
    strat_sample = stratified_sample(dataframe, str_variable, sample_size)
    strat_sample_counts = strat_sample.groupby(str_variable)[num_variable].count()
    strat_sample_means = strat_sample.groupby(str_variable)[num_variable].mean()
    strat_sample_vars = strat_sample.groupby(str_variable)[num_variable].apply(lambda x : np.var(x)/len(x) * ((strat_count.loc[x.name] - len(x))/(strat_count.loc[x.name] - 1)))
    pop_mean = sum([count*mean for mean, count in zip(strat_sample_means, strat_count)])/N
    pop_variance = sum([count**2*variance for count, variance in zip(strat_count, strat_sample_vars)])/N**2
    return pop_mean, pop_variance
```

```python
def stratified_conf_int(pop_mean, pop_var, prct):
    bottom = pop_mean - (norm.ppf(prct+((1-prct)/2)) * (np.sqrt(pop_var)))
    ceiling = pop_mean + (norm.ppf(prct+((1-prct)/2)) * (np.sqrt(pop_var)))
    return bottom, ceiling
```

```python
mean, variance = stratified_pop_estimator(rest_cuisine, ['value'], 'avg_cost_2', 500)
print(mean, variance)
```

```python
rest_cuisine['avg_cost_2'].mean()
```

```python
stratified_conf_int(mean, variance, 0.95)
```

```python
means = []
variances = []
sample_sizes = np.arange(100, 10000, 100)
for x in sample_sizes:
    mean, variance = stratified_pop_estimator(rest_cuisine, ['value'], 'avg_cost_2', x)
    means.append(mean)
    variances.append(variance)
    
fig, (ax1, ax2) = plt.subplots(1,2, figsize=[15,5])
ax1.plot(sample_sizes, means)
ax2.plot(sample_sizes, variances)
plt.setp(ax1, xlabel='Sample size', ylabel='Mean')
plt.setp(ax2, xlabel='Sample size', ylabel='Variance')
plt.show()
```
